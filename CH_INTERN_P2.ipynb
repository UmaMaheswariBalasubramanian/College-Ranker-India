## Task 2: Visualizing the Target Distribution

```python
bx.revenue.hist()
```

```python
bx['revenue'].plot(kind='hist',
                      figsize=(15, 5),
                      bins=50,
                      title='Distribution of Revenue')
plt.show()
print('Revenue has mean {:1.0f} and standard deviation {:1.0f}'.format(bx['revenue'].mean(), bx['revenue'].std())) 
```

Log transforms are a common way to deal with features or targets that are heavily skewed. Log transforms are also easy to interperet. For every increase of 1 in the log transform, we can say the revenue increased 10x. With the log transform we can see we've reduced the skew.

```python
bx['revenue_log'] = bx['revenue'].apply(np.log)
bx['revenue_log'].plot(kind='hist',
                      figsize=(15, 5),
                      bins=50,
                      title='Distribution of Log Revenue')
plt.show()
```

##### Overview of Features

```python
bx['budget'].plot(kind='hist',
                      figsize=(15, 5),
                      bins=50,
                      title='Distribution of Budget',
                      color='blue')
plt.show()

# Use the log1p transform since some values are zero
bx['budget_log'] = bx['budget'].apply(np.log)
bx['budget_log'] = bx['budget_log'].replace(-np.inf, 0)
bx['budget_log'].plot(kind='hist',
                      figsize=(15, 5),
                      bins=50,
                      title='Distribution of Log+1 Budget',
                      color='blue')
plt.show()
```

```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

sns.jointplot('budget_log', 'revenue_log', bx.loc[bx['budget_log'] > 1], kind='reg')
plt.show()
```

```python
bx['log_budget'] = np.log1p(bx['budget'])
bx['log_revenue'] = np.log1p(bx['revenue'])
```

```python
# Relationship between Film Revenue and Film Budget
import seaborn as s
plt.figure(figsize=(16,8))
plt.subplot(1,2,1)
s.scatterplot(bx['budget'], bx['revenue'])
plt.title('Budget v/s Revenue');
plt.subplot(1,2,2)
s.scatterplot(bx['log_budget'], bx['log_revenue'])
plt.title('Log Budget v/s Log Revenue');
```
